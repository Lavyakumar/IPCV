```python
img =cv2.imread('/content/noisy.png', 0 ) #grayscale0
```

```python
f= np.fft.fft2(img) #(H freq abhi center mai h) fft algo mai ff2 fn for 2D img to conver img into frequency spectrum
```

```python
fshift= np.fft.fftshift(f) # low mid mai aa gai
```

```python
row, col= img.shape
```

```python
crow, ccol= row//2, col//2
```

```python
mask_hp = np.ones((row, col), np.uint8)
```

```python
mask_hp[crow-1:crow+1,ccol-1:ccol+1]=0
```

```python
hp=np.fft.ifft2(np.fft.ifftshift(fshift*mask_hp))
```

```python
cv2.imwrite("noisy_hp1.png", np.abs(hp))
```

```python
files.download("noisy_hp1.png")
```

```python
mask_lp= np.ones((row, col), np.uint8)
```

```python
mask_lp[crow-30:crow+30,ccol-30:ccol+30]=1
```

```python
lp=np.fft.ifft2(np.fft.ifftshift(fshift*mask_lp))
```

```python
cv2.imwrite("noisy_LP1.png", np.abs(lp))
```
